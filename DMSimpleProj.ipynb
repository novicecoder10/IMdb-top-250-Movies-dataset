```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("sample_data/iMDbTop250Movies.csv",na_values='NA')
df.head
```

```python
df.plot.hist()
```

```python
df.boxplot(vert=False)
```

```python
df.describe(include='all')
df.plot()
df.plot(x='Title',y='imdb_rating')
df.plot(subplots=True, figsize=(8,8))
```

```python
plt.plot(df.imdb_rating,color='blue')
```

```python
plt.scatter(df.year,df.imdb_rating)
plt.xlabel("year")
plt.ylabel("Metascore")
plt.title('Scatter plot')
```

```python
plt.hist(df.Metascore)
```

```python
plt.bar(df.Title,df.imdb_rating)
```

```python
import seaborn as sns
sns.relplot(data=df,x='year',y='imdb_rating')
```

```python
sns.lmplot(data=df,x='year',y='imdb_rating',hue='Title')
```

```python
sns.pairplot(data=df,hue="Rating")
```

```python
sns.catplot(data=df,x='year',y='Metascore')
```
